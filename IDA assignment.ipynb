<h6>Q1.Find the car with the best mpg, worst mpg.<h6>

```python
maxmpg = data['mpg'].max()
print('MODEL WITH BEST MPG')
print(maxmpg)
a=data['mpg'].idxmax()
print(data['model'].loc[[a]])
print('\n' * 2)
print('#' * 20)
print('\n' * 2)
print('MODEL WITH WORST MPG')
minmpg = data['mpg'].min()
print(minmpg)
b=data['mpg'].idxmin()
print(data['model'].loc[[b]])
```

<H6> Q7.What is the correlation between the weight of the car and MPG? Draw a scatter plot. What is 
the relationship between the two attributes?<H6>
ANS. Fuel efficiency(MPG) and vehicle weight are highly negatively correlated with a correlation coeff. of -0.868

```python
plt.rcParams.update({'figure.figsize':(10,8), 'figure.dpi':100})
sns.lmplot(x='wt', y='mpg', data=data)
plt.title("Scatter Plot with Linear fit");
from scipy.stats import pearsonr
corr, _ = pearsonr(data['mpg'], data['wt'])
print('Pearsons correlation: %.3f' % corr)
```
